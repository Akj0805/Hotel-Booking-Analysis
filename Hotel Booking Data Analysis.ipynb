# Data Analysis & Visulization

```python
# Calculate ratio of cancellation of Booking 
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled','Canceled'], df['is_canceled'].value_counts(),edgecolor = 'k', width = 0.7)
plt.show()
```

```python
# Create a new figure for the count plot with a size of 8x4 inches
plt.figure(figsize=(8, 4))

# Create a count plot using Seaborn's `countplot` function
# Set the x-axis as 'hotel', the hue as 'is_canceled', and the data as the DataFrame 'df'
# Use the 'Blues' palette for coloring the bars
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Get the legend handles and labels for customizing the legend
legend_labels, _ = ax1.get_legend_handles_labels()

# Set the title for the count plot with a size of 20
plt.title('Reservation status in different hotels', size=20)

# Set the x-axis label as 'hotel' and the y-axis label as 'Number of Reservations'
plt.xlabel('hotel')
plt.ylabel('Number of Reservations')

# Display the count plot
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Create a new figure for the line plot with a size of 20x8 inches
plt.figure(figsize=(20, 8))

# Set the title for the line plot with a font size of 30
plt.title('Average Daily rate in city and resort hotel', fontsize=30)

# Plot the average daily rate for the resort hotel
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')

# Plot the average daily rate for the city hotel
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')

# Add a legend to the plot with a font size of 20
plt.legend(fontsize=20)

# Display the line plot
plt.show()
```

```python
# Extract the month from the 'reservation_status_date' column and assign it to a new 'month' column
df['month'] = df['reservation_status_date'].dt.month

# Create a new figure for the count plot with a size of 16x8 inches
plt.figure(figsize=(16, 8))

# Create a count plot using Seaborn's `countplot` function
# Set the x-axis as 'month', the hue as 'is_canceled', and the data as the DataFrame 'df'
# Use the 'bright' palette for coloring the bars
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

# Get the legend handles and labels for customizing the legend
legend_labels, _ = ax1.get_legend_handles_labels()

# Set the title for the count plot with a size of 20
plt.title('Reservation Status Per Month', size=20)

# Set the x-axis label as 'month' and the y-axis label as 'number of reservations'
plt.xlabel('month')
plt.ylabel('number of reservations')

# Add a legend with custom labels for the 'not_canceled' and 'canceled' categories
plt.legend(['not_canceled', 'canceled'])

# Display the count plot
plt.show()
```

```python
# Create a new figure for the bar plot with a size of 15x8 inches
plt.figure(figsize=(15, 8))

# Set the title for the bar plot with a font size of 30
plt.title('ADR per Month', fontsize=30)

# Filter the DataFrame to include only canceled reservations and group by month
# Calculate the sum of ADR for each month and reset the index to make it a DataFrame
grouped_data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Create a bar plot using Seaborn's `barplot` function
# Set the x-axis as 'month', the y-axis as 'adr', and the data as the grouped data
sns.barplot(x='month', y='adr', data=grouped_data)

# Add a legend to the plot with a font size of 20
plt.legend(fontsize=20)

# Display the bar plot
plt.show()
```

```python
# Filter the DataFrame to include only canceled reservations
cancelled_data = df[df['is_canceled'] == 1]

# Count the occurrences of each country in the canceled reservations and select the top 10 countries
top_10_country = cancelled_data['country'].value_counts()[:10]

# Create a new figure for the pie chart with a size of 8x8 inches
plt.figure(figsize=(8, 8))

# Set the title for the pie chart
plt.title('Top Countries with Reservation Cancelled')

# Create a pie chart with the top 10 countries, displaying the percentage of each country's occurrence
# The `autopct` parameter specifies the format of the percentage values
# The `labels` parameter specifies the labels for each pie slice, which are the country names
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)

# Display the pie chart
plt.show()
```

```python
# Group canceled reservations by reservation_status_date and calculate the average adr
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Filter not canceled reservations and calculate the average adr
not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the average daily rate for both canceled and not canceled reservations
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend()

# Display the plot
plt.show()
```

```python
# Filter the 'cancelled_df_adr' DataFrame based on date conditions
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016') &
                                    (cancelled_df_adr['reservation_status_date'] < '2017-09')]

# Filter the 'not_cancelled_df_adr' DataFrame based on date conditions
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016') &
                                            (not_cancelled_df_adr['reservation_status_date'] < '2017-09')]
```

```python
# Plot the average daily rate for both canceled and not canceled reservations
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend()

# Display the plot
plt.show()
```
